```python
## import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## read the csv file
df = pd.read_csv('DataAnalyst 2.csv')
df.columns = df.columns.str.replace(' ', '_')
df
```

```python
## clean the salary column
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace('$', '')
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace('K', '')
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace('(Glassdoor est.)', '')
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace('(', '')
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace(')', '')
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace(' ', '')
df['Salary_Estimate'] = df['Salary_Estimate'].str.split('-')
# df['Salary_Estimate'] = (df['Salary_Estimate'].str[0].astype(int) + df['Salary_Estimate'].str[1].astype(int))/2
df.head()
```

```python
## functions
# make the salary numeric
def salary(x):
    if len(x) == 2:
        if x[0] == '':
            return int(x[1])
        else:
            ans = (int(x[0]) + int(x[1])) / 2
            return ans
    else:
        return x

# identify if the job title is for senior or not
def senior(x):
    if 'senior' in x.lower():
        return 1
    else:
        return 0
```

```python
# applying the functions from above to the dagtaframe
df['Salary_Estimate'] = df['Salary_Estimate'].apply(lambda x: salary(x))
df['Senior'] = df['Job_Title'].apply(lambda x: senior(x))
df.head()
```

### Word count in the Job description

```python
import collections
box = ''

for row in df.itertuples():
    box += row.Job_Description
    box += ' '
box = box.split()

## 'c' has all the list word counts in the job description
c = collections.Counter(box)
print(c)

## 'and': 58630 means that the word 'and' showed up in the whole dataframe 58630 times
```

```python
## you can find words like a dictionary
print(c['MongoDB'])
```

```python

def words(x):
    x = x.lower()
    box = []
    box.append(x)
    box.append(x.upper())
    box.append(x.capitalize())
    ans = {}
    ans[x] = box
    return ans
```

```python
## list of hard skills
list_of_words = [
    'python','sql','R','SAS','SPSS','MATLAB','Spark','Hadoop','machine learning',
    'MongoDB','Tableau','PowerBI','Looker','Google Analytics','Casandra','Excel','Nosql'
]

ans = {}
count = 0
for w in list_of_words:
    word = words(w)
    for i in word[w.lower()]:
        count += c[i]
    ans[w] = count
    count = 0
print(ans)
## output is the hard skill word counts that we want to know
```
